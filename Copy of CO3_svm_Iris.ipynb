```python
# Import the necessary libraries
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import pandas as pd
```

```python

# Load a dataset (Iris dataset)
iris=datasets.load_iris()

# Extract fist two atributes as input
x=iris.data[:,:2]
y=iris.target

#Split training and test set
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python

# Create an SVM classifier (choose kernel functions- 'linear')
svm_classifier=SVC(kernel='linear',C=1,gamma=0.5)

# Train the SVM classifier on the training data
svm_classifier.fit(x_train,y_train)

# Make predictions on the test data
y_pred=svm_classifier.predict(x_test)

# Calculate accuracy
accuracy=accuracy_score(y_test,y_pred)

#Print Confusion Matrix
cm=confusion_matrix(y_test,y_pred)
print("Confusion matric\n",cm)
classification_report(y_test,y_pred)
```

```python
# Create an SVM classifier (choose kernel functions- 'poly')
svm_classifier=SVC(kernel='poly',degree=3,C=1.0,gamma='auto')

# Train the SVM classifier on the training data
svm_classifier.fit(x_train,y_train)

# Make predictions on the test data
y_pred=svm_classifier.predict(x_test)

# Calculate accuracy
accuracy=accuracy_score(y_test,y_pred)

#Print Confusion Matrix
cm=confusion_matrix(y_test,y_pred)
print("Confusion matric\n",cm)
classification_report(y_test,y_pred)
```
