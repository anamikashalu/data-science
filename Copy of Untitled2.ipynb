```python
import tensorflow as tf
from keras import datasets,layers,models
import matplotlib.pyplot as plt
```

```python
(train_images,train_labels),(test_images,test_labels)=datasets.mnist.load_data()
```

```python
print(train_images.shape)
print(test_images.shape)
```

```python
train_images,test_images=train_images/255.0,test_images/255.0
```

```python
model=models.Sequential([
    layers.Conv2D(6,(5,5),activation='relu',input_shape=(28,28,1)),
    layers.AveragePooling2D((2,2),strides=(2,2)),
    layers.Conv2D(16,(5,5))
])
```

```python
#compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#print a summary of the models architecture
model.summary()
```

```python
#train the model
learn=model.fit(train_images, train_labels, epochs=5, validation_data=(test_images, test_labels))
```

```python
#evaluate the model
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=0)
print (f"test_accuracy: {test_acc:.2f}")
```
