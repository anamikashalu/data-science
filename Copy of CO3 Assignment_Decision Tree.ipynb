Link of dataset :https://raw.githubusercontent.com/smitha-ks/Notebooks/Datasets/diabetes_dataset_DT.csv

Use Iris dataset to implement decision tree algorithm

```python
# Import the necessary libraries
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import pandas as pd


```

```python
# Load a dataset
df=pd.read_csv("https://raw.githubusercontent.com/smitha-ks/Notebooks/Datasets/diabetes_dataset_DT.csv")
```

```python
# Split the dataset into a training set and a testing set(trainig 66% records)
x=df.iloc[:,1:9]
y=df.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.37,random_state=1)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python

# Create a Decision Tree Classifier with gini index as criterion
DT_clf=DecisionTreeClassifier(random_state=0)
```

```python
# Train the classifier on the training data
DT_clf.fit(x_train,y_train)
```

```python

# Make predictions on the test data
y_pred=DT_clf.predict(x_test)

# Calculate accuracy and confusion matrix
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
cm=confusion_matrix(y_test,y_pred)
print("Confusion matric\n",cm)
classification_report(y_test,y_pred)

```

```python
#Plot decision tree
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plot_tree(DT_clf, filled=True,  rounded=True)
plt.show()
```
