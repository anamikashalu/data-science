Aim: Program to implement Na√Øve Bayes Algorithm using the given dataset named Naive-Bayes-Classification-Data.csv and find the accuracy of the algorithm. And also predict the given patients diabetic or not

The variable names in the dataset are as follows:

glucose,bloodpressure,diabetes(is patient diabetic or not?)


```python
#Import Modules
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
#Load dataset & do train_test_split
import pandas as pd
df = pd.read_csv("/content/diabetes - diabetes.csv")
x=df[['Glucose','BloodPressure']]
y=df['Outcome']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=2)
```

```python
#Implement Naive Bayes
gnb=GaussianNB()
gnb.fit(x_train,y_train)

```

Predict the values for unknown data
1. blood pressure =100, glucose=80
2. blood pressure =135, glucose=140


```python
#write code here
x_new=[[80,100]]
y_new=gnb.predict(x_new)
print("Predicted output for [[80,100]:",y_new)

x_new=[[140,135]]
y_new=gnb.fit(x_train,y_train).predict(x_new)
print("Predicted output for [[140,135]]:",y_new)
```
